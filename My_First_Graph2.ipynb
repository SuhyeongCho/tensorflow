```python
import tensorflow as tf
import numpy as np
```

```python
a = tf.placeholder(tf.int32,shape=[2],name="my_input")
b = tf.reduce_prod(a,name="prod_b")
c = tf.reduce_sum(a,name="sum_c")
d = tf.add(b,c,name="add_d")
```

```python
sess = tf.Session()
input_dict = {a:np.array([5,3],dtype=np.int32)}
sess.run(d)
```

```python
sess.close()
```
